```python
import pandas as pd

# Load the data
file_path = 'C:\\Users\\sreer\\OneDrive\\Desktop\\data analysis\\amazon.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data
print("First few rows of the dataset:")
print(data.head(), "\n")

# Basic information about the dataset
print("Dataset Info:")
print(data.info(), "\n")

# Check for missing values
print("Missing values in the dataset:")
print(data.isnull().sum(), "\n")

# Descriptive statistics of the dataset
print("Descriptive statistics:")
print(data.describe(), "\n")

# Analyzing specific columns
# Distribution of ratings
print("Rating distribution:")
print(data['rating'].value_counts(), "\n")

# Top 10 most reviewed products
print("Top 10 most reviewed products:")
data['rating_count'] = data['rating_count'].str.replace(',', '').astype(int)  # Convert rating_count to integer
top_reviewed = data.nlargest(10, 'rating_count')
print(top_reviewed[['product_name', 'rating_count']], "\n")

# Top 10 highest rated products
print("Top 10 highest rated products:")
data['rating'] = data['rating'].astype(float)  # Convert rating to float
top_rated = data.nlargest(10, 'rating')
print(top_rated[['product_name', 'rating']], "\n")

# Average discounted price
data['discounted_price'] = data['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
print("Average discounted price:")
print(data['discounted_price'].mean(), "\n")

# Average actual price
data['actual_price'] = data['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
print("Average actual price:")
print(data['actual_price'].mean(), "\n")

# Average discount percentage
data['discount_percentage'] = data['discount_percentage'].str.rstrip('%').astype(float)
print("Average discount percentage:")
print(data['discount_percentage'].mean(), "\n")
```

```python
import pandas as pd

# Load the data
file_path = 'C:\\Users\\sreer\\OneDrive\\Desktop\\data analysis\\amazon.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data
print("First few rows of the dataset:")
print(data.head(), "\n")

# Basic information about the dataset
print("Dataset Info:")
print(data.info(), "\n")

# Check for missing values
print("Missing values in the dataset:")
print(data.isnull().sum(), "\n")

# Descriptive statistics of the dataset
print("Descriptive statistics:")
print(data.describe(), "\n")

# Analyzing specific columns
# Distribution of ratings
print("Rating distribution:")
print(data['rating'].value_counts(), "\n")

# Top 10 most reviewed products
print("Top 10 most reviewed products:")
try:
    data['rating_count'] = data['rating_count'].str.replace(',', '').astype(int)  # Convert rating_count to integer
    top_reviewed = data.nlargest(10, 'rating_count')
    print(top_reviewed[['product_name', 'rating_count']], "\n")
except Exception as e:
    print("Error in processing 'rating_count':", e)

# Top 10 highest rated products
try:
    data['rating'] = data['rating'].astype(float)  # Convert rating to float
    top_rated = data.nlargest(10, 'rating')
    print(top_rated[['product_name', 'rating']], "\n")
except Exception as e:
    print("Error in processing 'rating':", e)

# Average discounted price
try:
    data['discounted_price'] = data['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
    print("Average discounted price:")
    print(data['discounted_price'].mean(), "\n")
except Exception as e:
    print("Error in processing 'discounted_price':", e)

# Average actual price
try:
    data['actual_price'] = data['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
    print("Average actual price:")
    print(data['actual_price'].mean(), "\n")
except Exception as e:
    print("Error in processing 'actual_price':", e)

# Average discount percentage
try:
    data['discount_percentage'] = data['discount_percentage'].str.rstrip('%').astype(float)
    print("Average discount percentage:")
    print(data['discount_percentage'].mean(), "\n")
except Exception as e:
    print("Error in processing 'discount_percentage':", e)

```
